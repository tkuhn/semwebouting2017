import the required packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import wget
```

download the data

```python
wget.download('https://vincentarelbundock.github.io/Rdatasets/csv/DAAG/cricketer.csv')
```

load CSV

```python
crick = pd.read_csv('cricketer.csv')
```

select columns

```python
crick[['year', 'life']]
```

individual columns, shorthand (tab autocompletes)

```python
crick.year
```

Select records 20 to 25

```python
crick.ix[20:25]
```

Plot one column against another

```python
plt.scatter(crick.year, crick.life)
plt.show()
```

Save the figure (and add some niceties)

```python
plt.scatter(crick.year, crick.life, alpha = 0.1, edgecolors='none')
plt.savefig('cricketers.pdf')
```

Filter out the cricketers that were still alive in 1992

```python
dc = crick[crick.dead == 1]
```

Plot and color by KIA

```python
plt.scatter(dc.year, dc.life, c=dc.kia, alpha=0.1, edgecolors='none')
plt.show()
```

Plot date of death vs. age

```python
plt.scatter(dc.year + dc.life, dc.life, c=dc.kia, alpha=0.5, edgecolors='none')
plt.xlabel('date of death')
plt.ylabel('date of death')
plt.savefig('result.png')
```
